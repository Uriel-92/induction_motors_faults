## Script for see the stratified labels

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv("labels_stratified.csv")
columns = {"Rotor-0":"Rotor", "Noload":"Healthy", "A&C&B10":"10%-stator 3-phase",
           "A50":"50%-stator 1-phase", "A&B50":"50%-stator 2-phase", "A&C30":"30%-stator 2-phase",
           "A&C10":"10%-stator 2-phase", "A10":"10%-stator 1-phase", "A&C&B30":"30%-stator 3-phase",
           "A30":"30%-stator 1-phase", "Fan":"Coling"}
df["Labels"] = df["Labels"].map(columns)
df["Labels"].value_counts().plot(kind="bar")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, _, _ = train_test_split(df["Labels"].values, test_size=0.2)
```
